```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
animals = pd.read_csv("data/animalia.csv", index_col="Common Name")
animals
```

plot the mass vs. length as a log-log scatter plot.

```python
animals.plot("Length / cm", "Mass / kg", kind="scatter", loglog=True)
```

table of animals with less than 10 kg

```python
animals[animals["Mass / kg"] < 10]
```

Collect all animals that have the same number of legs (use the .groupby() method) and make a bar graph of the maximum length for each group.

```python
legs = animals.groupby("Legs")
legs.max().plot(y="Length / cm", kind="bar")
plt.legend(loc="upper center")
```
